```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def df(x: np.ndarray):
    res = -2 * np.sin(x) * np.cos(x)
    return res


def f(x: float):
    res = math.cos(x)
    return res

def g(x: float):
    res = math.sin(x)
    return res

def g12(x: np.ndarray, indexX: int):
    res = g(x[indexX]) + g(x[indexX + 1])
    return res / 2
```

```python
minX = -10
maxX = 10
sizeX = 200
lengthX = maxX - minX
dx = lengthX / (sizeX - 1)

x = np.linspace(minX, maxX, sizeX)
dyExact = df(x)
dy = np.copy(dyExact)

plt.plot(x, dyExact)
```

```python
for i in range(1, sizeX - 1):
    dy[i] = f(x[i - 1]) * g12(x, i - 1) +\
        f(x[i]) * (-g12(x, i - 1) - g12(x, i)) +\
        f(x[i + 1]) * g12(x, i)
    dy[i] /= (dx * dx)

plt.plot(x, dy)
```

```python
error = np.abs(dy - dyExact)
plt.plot(x, error)
```
