```python
import numpy as np
import plotly.graph_objs as go
import pdb
from mpl_toolkits.mplot3d import Axes3D
```

```python
resultsPath = "../Results/"

filename = "surface.txt"
filepath = resultsPath + filename

sizeX = 0
sizeY = 0
sizeZ = 0

minX = 0
maxX = 0

minY = 0
maxY = 0

minZ = 0
maxZ = 0

t = 0

data = np.ndarray((0,0))

rowdata = np.genfromtxt(filepath, delimiter='\t',  skip_header=1)
dataSurface = rowdata[:, :-1]
```

```python
with open(filepath, 'r') as file:
    firstLine = file.readline()
    numbers = [int(number) for number in firstLine.strip().split()]

    sizeX = numbers[0]
    minX = numbers[1]
    maxX = numbers[2]

    sizeY = numbers[3]
    minY = numbers[4]
    maxY = numbers[5]

    t = numbers[6]
```

```python
x = np.linspace(minX, maxX, sizeX)
y = np.linspace(minY, maxY, sizeY)

Y, X = np.meshgrid(y, x)

fig = go.Figure(data=[go.Surface(z=dataSurface, x=X, y=Y)])
fig.update_layout(
    title='3D Surface plot',
    scene=dict(
        xaxis=dict(nticks=4, range=[minX, maxX],),
        yaxis=dict(nticks=4, range=[minY, maxY],),
        zaxis=dict(nticks=4, range=[np.min(dataSurface), np.max(dataSurface)],)))

fig.show()
```
