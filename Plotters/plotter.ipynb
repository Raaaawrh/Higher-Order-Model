```python
import numpy as np
import matplotlib.pyplot as plt
import pdb
from mpl_toolkits.mplot3d import Axes3D
```

```python
resultsPath = "../Results/"

filename = "surface.txt"
filepath = resultsPath + filename

sizeX = 0
sizeY = 0
sizeZ = 0

minX = 0
maxX = 0

minY = 0
maxY = 0

minZ = 0
maxZ = 0

t = 0

data = np.ndarray((0,0))

rowdata = np.genfromtxt(filepath, delimiter='\t',  skip_header=1)
data = rowdata[:, :-1]
```

```python
with open(filepath, 'r') as file:
    firstLine = file.readline()
    numbers = [int(number) for number in firstLine.strip().split()]

    sizeX = numbers[0]
    minX = numbers[1]
    maxX = numbers[2]

    sizeY = numbers[3]
    minY = numbers[4]
    maxY = numbers[5]

    t = numbers[6]
```

```python
x = np.linspace(minX, maxX, sizeX)
y = np.linspace(minY, maxY, sizeY)

Y, X = np.meshgrid(y, x)

fig = plt.figure(figsize=(10, 10))
axSufr = fig.add_subplot(111, projection='3d')
#axuVel = fig.add_subplot(1, 2, 2, projection='3d')
#axvVel = fig.add_subplot(2, 2, 1, projection='3d')

axSufr.plot_surface(X, Y, data)
```
